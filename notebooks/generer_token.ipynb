```
# Cellule 2 — Upload client_secrets.json
from google.colab import files
import os, shutil

os.makedirs('config', exist_ok=True)
print('📂 Upload ton fichier client_secrets.json :')
uploaded = files.upload()
for f in uploaded:
    shutil.move(f, 'config/client_secrets.json')
print('✅ Fichier prêt.')
```

```
# Cellule 3 — Cloner le repo et s'authentifier
import os, sys

GITHUB_REPO = 'https://github.com/Dodo/youtube-shorts-uploader.git'

# Fix : utiliser $GITHUB_REPO et non {GITHUB_REPO}
os.environ['GITHUB_REPO'] = GITHUB_REPO
!git clone $GITHUB_REPO

sys.path.insert(0, 'youtube-shorts-uploader/src')

from uploader import authenticate
youtube = authenticate('config/client_secrets.json', 'config/token.pickle')
print('✅ Authentification réussie !')
```

```
# Cellule 4 — Générer les secrets à copier dans GitHub
import base64

with open('config/token.pickle', 'rb') as f:
    token_b64 = base64.b64encode(f.read()).decode()

with open('config/client_secrets.json', 'r') as f:
    client_secrets = f.read()

print('=' * 60)
print('📋 SECRET 1 — Nom : CLIENT_SECRETS_JSON')
print('Valeur :')
print(client_secrets)
print()
print('=' * 60)
print('📋 SECRET 2 — Nom : YOUTUBE_TOKEN_PICKLE_B64')
print('Valeur :')
print(token_b64)
print('=' * 60)
print()
print('👆 Copie ces deux valeurs dans GitHub → Settings → Secrets')
```
