```
# Cellule 2 — Upload client_secrets.json
from google.colab import files
import os, shutil

os.makedirs('config', exist_ok=True)
print('📂 Upload ton fichier client_secrets.json :')
uploaded = files.upload()
for f in uploaded:
    shutil.move(f, 'config/client_secrets.json')
print('✅ Fichier prêt.')
```

```
# Cellule 3 — Télécharger uploader.py directement depuis GitHub (sans clone)
import os, sys

os.makedirs('src', exist_ok=True)

# Téléchargement direct du fichier raw (pas besoin de cloner)
!curl -s -o src/uploader.py "https://raw.githubusercontent.com/Dodo/youtube-shorts-uploader/main/src/uploader.py"

# Vérifier que le fichier est bien téléchargé
if os.path.getsize('src/uploader.py') > 100:
    print('✅ uploader.py téléchargé')
else:
    print('❌ Erreur — vérifie que ton repo GitHub est PUBLIC')

sys.path.insert(0, 'src')
```

```
# Cellule 4 — Authentification YouTube
from uploader import authenticate

youtube = authenticate('config/client_secrets.json', 'config/token.pickle')
print('✅ Authentification réussie !')
```

```
# Cellule 5 — Générer les secrets à copier dans GitHub
import base64

with open('config/token.pickle', 'rb') as f:
    token_b64 = base64.b64encode(f.read()).decode()

with open('config/client_secrets.json', 'r') as f:
    client_secrets = f.read()

print('=' * 60)
print('📋 SECRET 1 — Nom : CLIENT_SECRETS_JSON')
print('Valeur :')
print(client_secrets)
print()
print('=' * 60)
print('📋 SECRET 2 — Nom : YOUTUBE_TOKEN_PICKLE_B64')
print('Valeur :')
print(token_b64)
print('=' * 60)
print()
print('👆 Copie ces deux valeurs dans GitHub → Settings → Secrets')
```
